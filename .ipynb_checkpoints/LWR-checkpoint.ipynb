```python
import numpy as np
import matplotlib.pyplot as plt

tau = 0.5
X_train = np.array(range(3, 33)).reshape(-1, 1)
y_train = np.array([1, 2, 1, 2, 1, 1, 3, 4, 5, 4, 5, 6, 5, 6, 7, 8, 9, 10, 11, 11, 12, 11, 11, 10, 12, 11, 11, 10, 9, 8])
X_test = np.linspace(X_train.min(), X_train.max(), 400).reshape(-1, 1)
plt.scatter(X_train, y_train)
plt.show()

def locally_weighted_regression(X, y, x_query, tau):
    return [x0.dot(np.linalg.pinv(X.T.dot(np.diag(np.exp(-np.sum((X - x0) ** 2, axis=1) / (2 * tau ** 2)))).dot(X)).dot(X.T).dot(np.diag(np.exp(-np.sum((X - x0) ** 2, axis=1) / (2 * tau ** 2)))).dot(y)) for x0 in x_query]

y_test = locally_weighted_regression(X_train, y_train, X_test, tau)
plt.scatter(X_train, y_train, color='blue', label='Training Data')
plt.plot(X_test, y_test, color='red', label='Predictions')
plt.legend()
plt.show()
```
