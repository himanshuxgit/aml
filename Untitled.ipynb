```python
import numpy as np
import matplotlib.pyplot as plt
import random
tau=0.5
x_train=np.array(range(3,33)).reshape(-1,1)
y_train=np.array([random.randint(1,12) for _ in range(30)])
x_test=np.linspace(x_train.min(),x_train.max(),1000).reshape(-1,1)
def lwr(X, y, x_query, tau):
    return [x0.dot(np.linalg.pinv(X.T.dot(np.diag(np.exp(-np.sum((X - x0) ** 2, axis=1) / (2 * tau ** 2)))).dot(X)).dot(X.T).dot(np.diag(np.exp(-np.sum((X - x0) ** 2, axis=1) / (2 * tau ** 2)))).dot(y)) for x0 in x_query]            
y_test=lwr(x_train,y_train,x_test,tau)
plt.scatter(x_train,y_train,color='green',label='training')
plt.plot(x_test,y_test,color='red',label='Predictions')
plt.legend()
```

```python
import pandas as pd 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier,plot_tree
from sklearn.metrics import accuracy_score,confusion_matrix
from sklearn.preprocessing import LabelEncoder
df = pd.read_csv('Iris.csv')
print(df.head(7))
x,y=df.drop('Species',axis=1),LabelEncoder().fit_transform(df['Species'])
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2) 
dtree=DecisionTreeClassifier().fit(x_train,y_train)
y_pred=dtree.predict(x_test)
plot_tree(dtree,feature_names=df.columns[:-1],class_names=["setosa","vercicolor","virginica"])
```
