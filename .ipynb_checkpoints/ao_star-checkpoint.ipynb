```python
class Graph:
    def __init__(self, graph, heuristicNodeList, startNode):
        self.graph = graph
        self.H = heuristicNodeList
        self.start = startNode
        self.parent = {}
        self.status = {}
        self.solutionGraph = {}

    def applyAOStar(self):
        self.aoStar(self.start, False)
        print("FOR GRAPH SOLUTION, TRAVERSE THE GRAPH FROM THE STARTNODE:", self.start)
        print("------------------------------------------------------------")
        print(self.solutionGraph)
        print("------------------------------------------------------------")

    def aoStar(self, v, backTracking):
        print("HEURISTIC VALUES :", self.H)
        print("SOLUTION GRAPH :", self.solutionGraph)
        print("PROCESSING NODE :", v)
        print("-----------------------------------------------------------------------------------------")

        if v not in self.status or self.status[v] >= 0:
            minimumCost, childNodeList = 0, []
            for nodeInfoTupleList in self.graph.get(v, []):
                cost = sum(self.H.get(c, 0) + weight for c, weight in nodeInfoTupleList)
                if not childNodeList or cost < minimumCost:
                    minimumCost, childNodeList = cost, [c for c, _ in nodeInfoTupleList]

            self.H[v] = minimumCost
            self.status[v] = len(childNodeList)
            solved = all(self.status.get(childNode, 0) == -1 for childNode in childNodeList)
            if solved:
                self.status[v] = -1
                self.solutionGraph[v] = childNodeList

            if v != self.start:
                self.aoStar(self.parent.get(v, None), True)

            if not backTracking:
                for childNode in childNodeList:
                    self.parent[childNode] = v
                    self.status[childNode] = 0
                    self.aoStar(childNode, False)

# Example usage
h1 = {'A': 1, 'B': 6, 'C': 2, 'D': 12, 'E': 2, 'F': 1, 'G': 5, 'H': 7, 'I': 7, 'J': 1, 'T': 3}
graph1 = {
    'A': [[('B', 1), ('C', 1)], [('D', 1)]],
    'B': [[('G', 1)], [('H', 1)]],
    'C': [[('J', 1)]],
    'D': [[('E', 1), ('F', 1)]],
    'G': [[('I', 1)]]
}
G1 = Graph(graph1, h1, 'A')
G1.applyAOStar()
```
